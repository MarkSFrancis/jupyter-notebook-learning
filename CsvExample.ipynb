```python
import pandas
```

```python
sample_data = pandas.read_csv('sample-data.csv')
```

```python
sample_data
```

```python
def highlight_10k_and_over(value):
    styles = []
    
    for row in value:
        color = 'black'
        
        try:
            if isinstance(row, str):
                valueFloat = float(row.replace(',',''))
                if valueFloat >= 10000:
                    color = 'red'
        except ValueError:
            color = 'black'

        rowColor = 'color: {0}'.format(color)
        styles.append(rowColor)
    
    return styles
```

```python
sample_data.style.apply(highlight_10k_and_over, subset=['Population density (/km)'])
```
